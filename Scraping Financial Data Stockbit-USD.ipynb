```python
import selenium
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import NoSuchAttributeException
import time
import numpy as np
import pandas as pd
```

```python
# Openning chromedriver & stockbit
path_chrome = input('Enter path to webdriver Chrome')
driver = webdriver.Chrome(path_chrome)
driver.get('https://stockbit.com')
time.sleep(2)
driver.find_element_by_xpath('//*[@class="login-ldn"]').click()
path_file = input('Enter your path for file to save: ')
```

```python
# Login
loginID = input('Enter your stockbit login id: ')
password = input('Enter your stockbit password')
login_box = driver.find_element_by_name('username')
login_box.send_keys(loginID)
pass_box = driver.find_element_by_name('password')
pass_box.send_keys(password)
pass_box.send_keys(Keys.ENTER)
```

```python
# Select the company ticker
comp = 'HRUM'
driver.get('https://stockbit.com/#/symbol/'+ comp +'/financials')
```

```python
# Choosing the currency, period, language of financial data
# # Select Period
period = 'ANNUALLY'
se2 = Select(driver.find_element_by_xpath('//*[@class=" statement-type input-medium"]'))
se2.select_by_value(period)
time.sleep(2)

# Select Currency
currency = 'usd'
se3 = Select(driver.find_element_by_xpath('//*[@class=" currency_val input-mini"]'))
se3.select_by_value(currency)
time.sleep(2)

# Select language
language = '1' # 1 for English
se4 = Select(driver.find_element_by_xpath('//*[@class=" transType input-mini"]'))
se4.select_by_value(language)
time.sleep(2)
```

```python
 Select Report (BS, IS, or CF)
se = Select(driver.find_element_by_xpath('//*[@class="reportType input-medium"]'))
se.select_by_value('bs')

# Getting header of table
header = driver.find_elements_by_xpath('//table[@class="display fin-table dataTable"]/thead/tr/th')
header_text = [col.text for col in header]

# Click all expand button
for j in range(10):
    i = 0
    while True:
        elements = driver.find_elements_by_xpath('//*/i[@class="icon-right-dir collapsed"]')
        if len(elements) > i:
            try:
                elements[i].click() # click on the i-th element in the list
                i += 1 # increment i
                time.sleep(0.5) # wait until list will be updated
            except ElementClickInterceptedException:
                i += 1
            except ElementNotInteractableException:
                i += 1
            continue
        break
    
# Getting table body
data1_bs = []
elements = driver.find_elements_by_xpath('//table[@class="display fin-table dataTable"]/tbody/tr/td')
for element in elements:
    if element.get_attribute('data-raw') != None:                                        #Taking numbers data
        if element.get_attribute('data-raw') != '-':
            data1_bs.append(element.get_attribute('data-raw'))
        else:
            data1_bs.append(None)
    else:                                                                                #Taking account name data
        if element.text != '':
            if 'Total' in element.text and 'Total' not in element.find_element_by_class_name('acc-name').get_attribute('title'):
                data1_bs.append('Total ' + element.find_element_by_class_name('acc-name').get_attribute('title'))
            else:
                try:
                    data1_bs.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
                except NoSuchElementException:
                    data1_bs.append(element.text.strip())
        else:
            try:
                data1_bs.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
            except NoSuchElementException:
                data1_bs.append('')

data2_bs = np.array(data1_bs)
data3_bs = data2_bs.reshape(int(len(data1_bs)/len(header_text)),len(header_text))



# Select Report (BS, IS, or CF)
se = Select(driver.find_element_by_xpath('//*[@class="reportType input-medium"]'))
se.select_by_value('is')
time.sleep(2)

# Click all expand button
for j in range (10):
    i = 0
    while True:
        elements = driver.find_elements_by_xpath('//*/i[@class="icon-right-dir collapsed"]')
        if len(elements) > i:
            try:
                elements[i].click() # click on the i-th element in the list
                i += 1 # increment i
                time.sleep(0.5) # wait until list will be updated
            except ElementClickInterceptedException:
                i += 1
            except ElementNotInteractableException:
                i += 1
            continue
        break


#Getting table body
data1_ist = []
elements = driver.find_elements_by_xpath('//table[@class="display fin-table dataTable"]/tbody/tr/td')
for element in elements:
    if element.get_attribute('data-raw') != None:                                        #Taking numbers data
        if element.get_attribute('data-raw') != '-':
            data1_ist.append(element.get_attribute('data-raw'))
        else:
            data1_ist.append(None)
    else:                                                                                #Taking account name data
        if element.text != '':
            if 'Total' in element.text and 'Total' not in element.find_element_by_class_name('acc-name').get_attribute('title'):
                data1_ist.append('Total ' + element.find_element_by_class_name('acc-name').get_attribute('title'))
            else:
                try:
                    data1_ist.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
                except NoSuchElementException:
                    data1_ist.append(element.text.strip())
        else:
            try:
                data1_ist.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
            except NoSuchElementException:
                data1_ist.append('')

data2_ist = np.array(data1_ist)
data3_ist = data2_ist.reshape(int(len(data1_ist)/len(header_text)),len(header_text))



# Select Report (BS, IS, or CF)
se = Select(driver.find_element_by_xpath('//*[@class="reportType input-medium"]'))
se.select_by_value('cf')
time.sleep(2)


# Click all expand button
for j in range(10):
    i = 0
    while True:
        elements = driver.find_elements_by_xpath('//*/i[@class="icon-right-dir collapsed"]')
        if len(elements) > i:
            try:
                elements[i].click() # click on the i-th element in the list
                i += 1 # increment i
                time.sleep(0.5) # wait until list will be updated
            except ElementClickInterceptedException:
                i += 1
            except ElementNotInteractableException:
                i += 1
            continue
        break


#Getting table body
data1_cf = []
elements = driver.find_elements_by_xpath('//table[@class="display fin-table dataTable"]/tbody/tr/td')
for element in elements:
    if element.get_attribute('data-raw') != None:                                        #Taking numbers data
        if element.get_attribute('data-raw') != '-':
            data1_cf.append(element.get_attribute('data-raw'))
        else:
            data1_cf.append(None)
    else:                                                                                #Taking account name data
        if element.text != '':
            if 'Total' in element.text and 'Total' not in element.find_element_by_class_name('acc-name').get_attribute('title'):
                data1_cf.append('Total ' + element.find_element_by_class_name('acc-name').get_attribute('title'))
            else:
                try:
                    data1_cf.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
                except NoSuchElementException:
                    data1_cf.append(element.text.strip())
        else:
            try:
                data1_cf.append(element.find_element_by_class_name('acc-name').get_attribute('title'))
            except NoSuchElementException:
                data1_cf.append('')

data2_cf = np.array(data1_cf)
data3_cf = data2_cf.reshape(int(len(data1_cf)/len(header_text)),len(header_text))
```

```python
bs = pd.DataFrame(data3_bs,columns=header_text)
ist = pd.DataFrame(data3_ist,columns=header_text)
cf = pd.DataFrame(data3_cf,columns=header_text)
```

```python
# Function to change number from strings format to integer
def change_to_num_mill(x):
    if x not in ('',None,np.nan):
        try:
            return int(float(x))/1000000
        except ValueError:
            return x/1000000
    else :
        return np.nan
```

```python
# Apply change format function to the dataframe
for col in header_text[1:]:
    bs[col] = bs[col].apply(lambda x: change_to_num_mill(x)).round(0)
    ist[col] = ist[col].apply(lambda x: change_to_num_mill(x)).round(0)
    cf[col] = cf[col].apply(lambda x: change_to_num_mill(x)).round(0)
```
